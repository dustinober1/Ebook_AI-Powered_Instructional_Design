# Lab: AI Prompt Engineering Playground

Welcome to your first interactive lab! In this notebook, we will explore how changing a prompt can drastically change the output of an AI model.

Since we don't have a live API key in this demo, we will simulate the responses to show you the concept.

```python
def simulate_llm_response(prompt):
    """Simulates an LLM response based on keywords in the prompt."""
    prompt_lower = prompt.lower()
    
    if "learning objective" in prompt_lower:
        if "bloom" in prompt_lower:
            return "Generated Objective: By the end of this module, learners will be able to ANALYSIS the impact of AI on ID workflows."
        else:
            return "Generated Objective: Understand AI in ID."
    elif "quiz" in prompt_lower:
        return "1. What is the primary benefit of AI? (a) Speed (b) cost (c) Accuracy"
    else:
        return "I am not sure what you want me to generate. Try asking for a learning objective."
```

## Task 1: Basic Prompting
Let's try a simple, vague prompt.

```python
simple_prompt = "Write a learning objective for this course."
print(simulate_llm_response(simple_prompt))
```

## Task 2: Advanced Prompting
Now let's add constraints. We want the objective to be aligned with **Bloom's Taxonomy**.

```python
advanced_prompt = "Write a learning objective using Bloom's Taxonomy."
print(simulate_llm_response(advanced_prompt))
```

### Conclusion
Notice how the specific instruction changed the quality of the output? This is the core of prompt engineering.

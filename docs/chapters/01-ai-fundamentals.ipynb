## Exercise: Simple Text Analysis

Below we use a simple list of keywords to identify 'Technical' vs 'Non-Technical' content snippets.

```python
snippets = ["Learn Python basics", "Introduction to Pedagogy"]
tech_keywords = ["Python", "Code", "Data"]

for s in snippets:
    category = "Technical" if any(k in s for k in tech_keywords) else "Non-Technical"
    print(f"Snippet: '{s}' -> Category: {category}")
```
